# 定义一个门票系统
- 构造函数
    - 门票原价是100元
    - 当周末的时候门票涨价20%
    - 小孩子半票
- 功能函数
    - 计算2个成人和1个小孩平日的票价

```python
class Ticket(object):
    def __init__(self, weekend=False, child=False):
        self.exp = 100
        if weekend:
            self.inc = 1.2
        else:
            self.inc = 1
        
        if child:
            self.discount = 0.5
        else:
            self.discount = 1
            
    def cal_price(self, num):
        return self.exp * self.inc * self.discount * num
# 对象实例化大人和小孩
adult = Ticket()
child = Ticket(child=True)

price = adult.cal_price(2) + child.cal_price(1)
print('两个成年人和一个小孩平日的价格时{}'.format(price))
```

# 游戏编程：按以下要求定义一个乌龟类和鱼类并尝试编程
- 假设游戏场景为范围（x, y)为 0<=x<=10, 0<=y<=10
- 游戏生成一只乌龟和10条鱼
- 他们的移动方向均随机
- 乌龟的最大移动能力是2（乌龟可以随机选择移动1还是2），鱼的最大移动能力是1
- 当移动到场景边缘，自动向反方向移动
- 乌龟初始化体力为100（上限）
- 乌龟每移动一次，体力消耗1
- 当乌龟和鱼重叠，乌龟吃掉鱼，乌龟体力增加20
- 鱼不计算体力
- 当乌龟体力为0或者鱼的数量为0时，游戏结束

```python
import tkinter
import random


class Tortoise(object):
    def __init__(self):
        # 设置乌龟的体力
        self.power = 100
        
        # 初始化乌龟的位置
        self.x = random.randint(0, 10)
        self.y = random.randint(0, 10)
        
        self.xv = random.choice([1, 2])
        self.yv = random.choice([1, 2])
        
    def move(self):
        self.x = self.x + self.xv
        self.y = self.y + self.yv

        # 判断乌龟是否超出边界
        if self.x < 0:
            self.x = -self.x
        if self.x > 10:
            self.x =  10 - self.x
        if self.y < 0:
            self.y = -self.y
        if self.y > 10:
            self.y = 10 - self.y

        self.power -= 1
        return(self.x, self.y)
             
    def eat(self):
        self.power += 20
        if self.power >= 100:
            self.power = 100

class Fish(object):
    def __init__(self):
        self.x = random.randint(0,10)
        self.y = random.randint(0,10)
        
        self.xv = 1
        self.yv = 1
        
    def move(self):
        self.x = self.x + self.xv
        self.y += self.yv
        
    # 判断鱼是否超出边界
        if self.x < 0:
            self.x = -self.x
        if self.x > 10:
            self.x =  10 - self.x
        if self.y < 0:
            self.y = -self.y
        if self.y > 10:
            self.y = 10 - self.y

        return(self.x, self.y)
    
class Screen(object):
    tortoises = list()
    fishs = list()
    
    def __init(self):
        self.num_fish = random.randint(6,10)
        self.root = tkinter.Tk()
        self.root.wm_title('龟鱼游戏')
        
        self.canvas = tkinter.Canvas(self.root, width=10, height=10)
        self.canvas.pack()
        for i in range(self.num_fish):
            fish = Fish()
            fish.move()
            self.fishs.append(fish)
            
        if not len(fishs):
            print('鱼已经被吃完了，游戏结束')
            
        if not tortoise.power():
            print('乌龟没有体力，游戏结束')
        
        self.meet()
            
        root.mainloop()
            
    def meet(self):
        pos = tortoise.move()
        
        # 在迭代器中做列表删除元素操作非常危险，经常会出现问题
        # 因为迭代器是直接引用列表元素做的操作
        # 所有我们这里用[:]将列表拷贝一份，然后再对列表操作
        for each_fish in fishs[:]:
            if each_fish.move() == pos:
                tortoise.eat
                fishs.remove(each.fish)
                print('有一个鱼被吃掉了')
    
    
if __name__ == '__main__':
    Screen()
```

# 定义一个点point和直线line类，使用get_len方法获取两点构成直线的长度
## 点类
- 构造函数
    - 初始化x，y坐标
- 功能函数
    - 返回x，y坐标
    
## 线类
- 构造函数
    - 初始化分别求出两个点x和y的差值
    - 通过勾股定理算出长度
- 功能函数
    - 返回算出的线的长度

```python
import math
class Point(object):
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def get_x(self):
        return self.x
    
    def get_y(self):
        return self.y
    
class Line(object):
    def __init__(self, p1, p2):
        self.x = p1.get_x() - p2.get_x()
        self.y = p1.get_y() - p2.get_y()
        
        self.len = math.sqrt(self.x * self.x + self.y * self.y)
        
    def get_len(self):
        return self.len
    
p1 = Point(3,4)
p2 = Point(6,8)
line = Line(p1,p2)
print(line.get_len())
```
